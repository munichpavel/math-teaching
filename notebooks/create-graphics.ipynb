```python
from pathlib import Path
import os

# Requires non-python, system dependency Graphviz: https://www.graphviz.org
from graphviz import Digraph, Source
```

```python
generated_graphics_dir = Path(os.environ['GENERATED_GRAPHICS_DIR'])
```

## Ways to allocate cards to 4 players, where each player only gets cards of a single suit

```python
dot_str = '''
digraph G {
    rankdir="LR"
    start -> "4 choices" [label="Select suit\nfor player 1"];
    "4 choices" -> "3 choices" [label="Select suit\nfor player 2"];
    "3 choices" -> "2 choices" [label="Select suit\nfor player 3"];
    "2 choices" -> "1 choice" [label="Select suit\nfor player 4"];
}
'''
pick_player_suits = Source(dot_str)
pick_player_suits.render(generated_graphics_dir / 'pick-player-suit', format='png', view=False)
pick_player_suits
```

## Above, but now the seating order of the players matters, too

```python
dot_str = '''
digraph G {
    rankdir="LR"
    chair_choice_4 [label="4 choices"];
    card_choice_4 [label="4 choices"];
    card_choice_3 [label="..."]
    
    start -> chair_choice_4 [label="Select player\nfor seat 1"];
    chair_choice_4 -> "3 choices" [label="Select player\nfor seat 2"];
    "3 choices" -> "2 choices" [label="Select player\nfor seat 3"];
    "2 choices" -> "1 choice" [label="Select player\nfor seat 4"];
    "1 choice" -> card_choice_4 [label="Select suit\nfor player in seat 1"];
    card_choice_4 -> card_choice_3 [label="Select suit\nfor player in seat 2"];
}
'''
pick_player_order_suits = Source(dot_str)
pick_player_order_suits.render(generated_graphics_dir / 'pick-player-order-suit', format='png', view=False)
pick_player_order_suits
```
